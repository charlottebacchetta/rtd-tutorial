## PART 1: Dataset 1 Analysis
This section focuses on analyzing **NYPD Arrest Data**. The goal is to clean the data, compute borough-wise summaries, and document the dataset.

### A) Load Dataset 1
1. Load the dataset and inspect its structure.
2. Select relevant columns and discard the rest.
3. Rename columns for clarity.

```python
import pandas as pd

# Step 1: Load Dataset 1
# This code reads the arrest dataset into a pandas DataFrame.
# The dataset contains records of arrests in NYC, including details about the arrest date, borough, 
# demographic information of the perpetrator, and the category of the offense.
df1 = pd.read_csv("/Users/charlottebacchetta/Desktop/Columbia/Computing in Context - U6006/Project 2/NYPD_Arrest_Data__Year_to_Date__20241122.csv")

# Step 2: Inspect the Data
# Display the first few rows of the DataFrame to verify the structure of the dataset and understand its contents.
df1.head()  # Provides a preview of the first few rows of the data.

# Print the number of rows (keys) and columns before filtering
# This step helps us understand the initial size of the dataset.
print(f"The dataset originally contains {df1.shape[0]} rows (keys) and {df1.shape[1]} columns.")

# Step 3: Keep Relevant Columns
# Select a subset of columns that are relevant to the analysis. Irrelevant columns are dropped to simplify the dataset.
# The selected columns include:
# - ARREST_DATE: The date of the arrest.
# - ARREST_BORO: The borough where the arrest occurred.
# - AGE_GROUP, PERP_SEX, PERP_RACE: Demographic information about the perpetrator.
# - LAW_CAT_CD: The category of the offense.
columns_to_keep = ['ARREST_DATE', 'ARREST_BORO', 'AGE_GROUP', 'PERP_SEX', 'PERP_RACE', 'LAW_CAT_CD']
df1 = df1[columns_to_keep]

# Print the number of rows and columns after filtering
# This step confirms that only the necessary columns remain.
print(f"After selection, the dataset contains {df1.shape[1]} columns.")

# Step 4: Rename Columns for Clarity
# Rename columns to more descriptive names that align with their contents. This makes the dataset easier to understand.
df1.rename(columns={
    'ARREST_DATE': 'Date',             # Date of the arrest
    'ARREST_BORO': 'Borough',          # Borough where the arrest occurred
    'AGE_GROUP': 'Age Group',          # Age group of the perpetrator
    'PERP_SEX': 'Sex',                 # Sex of the perpetrator
    'PERP_RACE': 'Race',               # Race of the perpetrator
    'LAW_CAT_CD': 'Level of offense'   # Category of the offense
}, inplace=True)

# Step 5: Display Cleaned Data
# Display the first few rows of the cleaned DataFrame to verify the changes.
df1.head()  # This ensures the dataset is structured correctly and ready for analysis.
```

### B) Handle Missing Values
1. Check for missing values.
2. Drop rows with missing values to clean the dataset.

```python
# Step 6: Check for Missing Values
# Before proceeding with analysis, it's essential to identify if there are any missing values in the dataset.
# Missing values can lead to errors or inaccuracies during analysis.
# The `.isnull()` function checks for null (missing) values in the DataFrame.
# The `.sum()` function aggregates the count of missing values for each column.
print("Missing values per column before cleaning:")
print(df1.isnull().sum())

# Step 7: Drop Rows with Missing Values
# Rows containing missing values are removed from the dataset using `.dropna()`.
# This ensures that all subsequent analysis is based on complete data.
# The `inplace=True` argument modifies the DataFrame in place, eliminating the need to reassign it.
df1.dropna(inplace=True)

# Step 8: Verify Changes
# Display the first few rows of the cleaned DataFrame to ensure that rows with missing values have been removed.
# Additionally, check for missing values again to confirm there are no null entries remaining.
print("Missing values per column after cleaning:")
print(df1.isnull().sum())

# Display the cleaned dataset
df1.head()
```

### C) Calculate Borough-Wise Mode
We calculate the mode for each column to summarize common characteristics for each borough.

```python
# Step 9: Define Columns for Mode Calculation
# These columns contain categorical data for which the mode (most frequent value) is meaningful.
# The mode provides insights into the most common demographic group or offense level in each borough.
columns_for_mode = ['Age Group', 'Sex', 'Race', 'Level of offense']

# Step 10: Group Data by Borough and Calculate the Mode
# The dataset is grouped by the 'Borough' column, ensuring calculations are performed for each borough separately.
# The `.agg()` function applies a custom lambda function to calculate the mode for each column.
# The lambda function handles cases where no mode exists by returning None, ensuring the process is robust.
borough_modes = df1.groupby('Borough')[columns_for_mode].agg(
    lambda x: x.mode()[0] if not x.mode().empty else None  # Get the first mode if available
)

# Step 11: Display Borough-Wise Modes
# The resulting DataFrame contains the mode for each specified column, grouped by borough.
# Each row represents a borough, and each column shows the most frequent value for the corresponding attribute.
print("Borough-Wise Modes Across Time:")
print(borough_modes)
```

### D) Data Dictionary
A data dictionary helps document the meaning of each column.

```python
# Step 12: Create a Data Dictionary
# A data dictionary provides a clear and concise description of each column in the dataset.
# This is especially useful for documentation and ensuring anyone working with the data understands its meaning.

data_dict1 = {
    'Date': 'Date of the arrest',               # Represents the date when the arrest occurred
    'Borough': 'Borough where the arrest took place',  # Indicates the borough where the arrest happened (e.g., Manhattan)
    'Age Group': 'Age group of the person arrested',   # Provides age demographic information of the arrested individual
    'Sex': 'Sex of the person arrested',               # Indicates the sex of the individual (e.g., Male or Female)
    'Race': 'Race of the person arrested'              # Represents the racial identity of the individual (e.g., Black, White)
}

# Display the data dictionary to ensure it is clear and complete
data_dict1
```

## PART 2: Dataset 2 Analysis
This section focuses on analyzing **Borough Community District Report**. We clean the data and compute borough-specific metrics.

### A) Load Dataset 2
1. Load the dataset and inspect its structure.
2. Select relevant columns and discard the rest.
3. Rename columns for clarity.

```python
import pandas as pd

# Step 1: Load Dataset 2
# This code reads the second dataset into a pandas DataFrame.
# The dataset includes information about SNAP, Cash Assistance, and Medicaid enrollees across NYC boroughs.
# The file path points to the location of the dataset on your system.
df2 = pd.read_csv("/Users/charlottebacchetta/Desktop/Columbia/Computing in Context - U6006/Project 2/Borough_Community_District_Report_20241122.csv")

# Step 2: Inspect the Data
# Display the first few rows of the dataset and its column names to understand its structure.
# This step verifies the data has been loaded correctly and helps identify the columns available for analysis.
df2.head(), df2.columns

# Step 3: Select Relevant Columns
# A subset of columns is chosen to focus on data relevant to the analysis.
# Irrelevant or redundant columns are excluded to simplify the dataset.
# Columns chosen include:
# - Month: The time period (monthly) for the data.
# - Borough: The NYC borough to which the data pertains.
# - Borough Consultation Total SNAP Recipients: Total individuals receiving SNAP benefits.
# - Borough Consultation Total Cash Assistance Recipients: Total individuals receiving cash assistance.
# - Borough Consultation Total Medicaid Only Enrollees: Individuals enrolled in Medicaid but not receiving other benefits.
columns_to_keep = [
    'Month', 
    'Borough', 
    'Borough Consultation Total SNAP Recipients', 
    'Borough Consultation Total Cash Assistance Recipients', 
    'Borough Consultation Total Medicaid Only Enrollees'
]
df2 = df2[columns_to_keep]

# Step 4: Rename Columns for Clarity
# Rename columns to more descriptive names that better reflect their content.
# Clear column names improve code readability and ease of use during analysis.
df2.rename(columns={
    'Month': 'Date',  # Rename 'Month' to 'Date' for consistency with the first dataset
    'Borough Consultation Total SNAP Recipients': 'Number of SNAP Benefit Recipients',
    'Borough Consultation Total Cash Assistance Recipients': 'Number of Cash Assistance Recipients',
    'Borough Consultation Total Medicaid Only Enrollees': 'Medicaid Enrollees'
}, inplace=True)

# Step 5: Verify Changes
# Display the first few rows of the cleaned dataset to confirm the column selection and renaming.
df2.head()
```

### B) Handle Missing Values
1. Check for missing values.
2. Drop rows with missing values to clean the dataset.

```python
# Step 6: Check for Missing Values
# Before analysis, it's essential to check for missing values in the dataset.
# Missing data can impact calculations and lead to errors or biases in the analysis.
# The `.isnull()` function identifies missing values in each column, and `.sum()` aggregates the count of missing values.
print("Missing values per column before cleaning:")
print(df2.isnull().sum())

# Step 7: Drop Rows with Missing Values
# Rows containing any missing values are removed using `.dropna()`.
# This ensures the dataset is complete and consistent for analysis.
# The `inplace=True` argument modifies the DataFrame directly without creating a copy.
df2.dropna(inplace=True)
```

### C) Calculate Borough-Wise Mean and Rank
Calculate the mean for relevant columns and rank boroughs in ascending order.

```python
# Step 10: Calculate Borough-Wise Mean Values
# The dataset is grouped by 'Borough', and the mean is calculated for all numerical columns.
# This provides borough-level averages for SNAP recipients, Cash Assistance recipients, and Medicaid enrollees.
# Grouping by borough helps compare how the average number of recipients varies across NYC boroughs.
borough_means = df2.groupby('Borough').mean()

# Step 11: Display Mean Values
# The result is a DataFrame where each row corresponds to a borough, 
# and each column shows the mean value of the corresponding numerical column for that borough.
print("Borough-Wise Mean Values:")
print(borough_means)

# Step 12: Define Columns for Ranking Boroughs
# The columns specified here will be used to rank boroughs based on their mean values.
# This helps in identifying boroughs with higher or lower average numbers of recipients for each category.
columns_to_rank = [
    'Number of SNAP Benefit Recipients', 
    'Number of Cash Assistance Recipients', 
    'Medicaid Enrollees'
]

# Step 13: Rank Boroughs by Each Metric
# A new DataFrame is created to rank boroughs for each of the specified columns.
# The `rank()` method assigns ranks based on the mean values (ascending or descending as needed).
borough_ranks = borough_means[columns_to_rank].rank(ascending=False)  # Higher values get higher ranks
```

### D) Data Dictionary
A data dictionary helps document the meaning of each column.

```python
# Step 14: Create a Data Dictionary for Dataset 2
# A data dictionary is created to describe the purpose and meaning of each column in the dataset.
# This is critical for understanding the structure of the dataset and ensures clarity for anyone working with the data.

data_dict2 = {
    'Date': 'Month and year of data collection',  # Time period when the data was collected
    'Borough': 'Name of the borough in New York City (e.g., Manhattan, Brooklyn, Queens)',  # Geographic identifier
    'Number of SNAP Benefit Recipients': 'The number of individuals in the borough receiving SNAP (Supplemental Nutrition Assistance Program) benefits',
    'Number of Cash Assistance Recipients': 'The number of individuals in the borough receiving cash assistance benefits',
    'Medicaid Enrollees': 'The number of individuals in the borough enrolled in Medicaid but not receiving other benefits'
}

# Step 16: Display the Updated Data Dictionary
# Loop through the dictionary and print each column name along with its description.
# This makes the data dictionary user-friendly and serves as clear documentation of the dataset.
print("Data Dictionary for Dataset 2:")
for column, description in data_dict2.items():
    print(f"{column}: {description}")
```

## PART 3: Merging Datasets and Analysis
This section integrates both datasets, ensures consistency, and analyzes trends.

### A) Data Synchronization and Merging
1. Check for time/date consistency between datasets.
2. Keep rows where data exists in both datasets.
3. Merge the datasets.

```python
# Step 1: Ensure 'Date' Columns are in the Same Format
# Convert the 'Date' columns in both datasets to datetime format for consistency.
# The `errors='coerce'` argument ensures that invalid date entries are converted to NaT (Not a Time).
df1['Date'] = pd.to_datetime(df1['Date'], errors='coerce')
df2['Date'] = pd.to_datetime(df2['Date'], errors='coerce')

# Step 2: Standardize 'Borough' Names for Consistency
# Standardize the 'Borough' column in both datasets by stripping whitespace and converting to uppercase.
# This ensures that variations like " Manhattan" and "MANHATTAN" are treated as identical values.
df1['Borough'] = df1['Borough'].str.strip().str.upper()
df2['Borough'] = df2['Borough'].str.strip().str.upper()

# Step 3: Identify and Log Mismatches in 'Borough' Values
# Print the unique values in the 'Borough' column for both datasets.
# This helps identify any discrepancies in naming conventions between the two datasets.
print("Unique Borough values in Dataset 1 (df1):", df1['Borough'].unique())
print("Unique Borough values in Dataset 2 (df2):", df2['Borough'].unique())

# Step 4: Optional - Resolve Borough Name Mismatches
# If discrepancies exist (e.g., abbreviations or alternative spellings), manually map these to a consistent format.
borough_mapping = {'M': 'MANHATTAN', 'B': 'BROOKLYN', 'Q': 'QUEENS', 'S': 'STATEN ISLAND', 'X': 'BRONX'}
df1['Borough'] = df1['Borough'].replace(borough_mapping)

# Step 5: Merge the Datasets
# Merge the two datasets on 'Borough' and 'Date' columns using an inner join.
# An inner join ensures only rows with matching 'Borough' and 'Date' values in both datasets are included.
merged_df = pd.merge(df1, df2, on=['Borough', 'Date'], how='inner')

# Step 6: Check for Missing Values in the Merged Dataset
# After merging, check for any missing values in the resulting dataset.
# This step ensures the integrity of the merged dataset.
missing_values = merged_df.isnull().sum()
print("\nMissing values after merging:")
print(missing_values)

# Step 7: Display the First Few Rows of the Merged Dataset
# Use `.head()` to preview the merged dataset and confirm it aligns with expectations.
print("\nMerged Dataset Preview:")
print(merged_df.head())
```

### B) Analyze Trends by Borough
1. Explore borough-specific trends over time.
2. Investigate relationships between variables.

```python
import matplotlib.pyplot as plt

# Step 1: Ensure 'Date' Column is in Monthly Format
# Convert the 'Date' column to monthly periods and back to timestamps for uniformity.
# This step standardizes the temporal granularity to monthly intervals, simplifying trend analysis.
merged_df['Month'] = merged_df['Date'].dt.to_period('M').dt.to_timestamp()

# Step 2: Group Data by Borough and Month
# Aggregate the data to calculate:
# - Total SNAP, Cash Assistance, and Medicaid recipients (sum) by borough and month.
# - Number of arrests (count of rows) by borough and month.
grouped_data = merged_df.groupby(['Borough', 'Month']).agg({
    'Number of SNAP Benefit Recipients': 'sum',
    'Number of Cash Assistance Recipients': 'sum',
    'Medicaid Enrollees': 'sum',
    'Date': 'count'  # Count rows to represent the number of arrests
}).rename(columns={'Date': 'Number of Arrests'})

# Step 3: Line Charts for SNAP, Cash Assistance, and Medicaid Recipients by Borough
# Generate one line chart per borough to visualize trends in benefits over time.
for borough in merged_df['Borough'].unique():
    # Extract data for the specific borough
    subset = grouped_data.loc[borough].reset_index()

    # Create a line chart
    plt.figure(figsize=(14, 8))
    plt.plot(subset['Month'], subset['Number of SNAP Benefit Recipients'], label='SNAP Benefit Recipients', marker='o')
    plt.plot(subset['Month'], subset['Number of Cash Assistance Recipients'], label='Cash Assistance Recipients', marker='o')
    plt.plot(subset['Month'], subset['Medicaid Enrollees'], label='Medicaid Enrollees', marker='o')

    # Configure the chart
    plt.title(f'Monthly Trends for {borough}: SNAP, Cash Assistance, and Medicaid', fontsize=14)
    plt.xlabel('Month', fontsize=12)
    plt.ylabel('Number of Individuals', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.legend()
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

# Step 4: Bar Charts for Number of Arrests by Borough
# Create a separate bar chart per borough showing monthly arrest counts.
for borough in merged_df['Borough'].unique():
    # Extract data for the specific borough
    subset = grouped_data.loc[borough].reset_index()

    # Create a bar chart
    plt.figure(figsize=(14, 8))
    plt.bar(subset['Month'].dt.strftime('%Y-%m'), subset['Number of Arrests'], color='skyblue', alpha=0.8)

    # Configure the chart
    plt.title(f'Number of Arrests per Month in {borough}', fontsize=14)
    plt.xlabel('Month', fontsize=12)
    plt.ylabel('Number of Arrests', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
```

### C) Compute Relevant Metrics
1. Calculate metrics to derive borough-specific insights.
2. Summarize and display results.
### Justification for the Metric
The 'Arrests per Recipient' metric was chosen to understand the proportionality of arrests to individuals 
receiving assistance. By normalizing arrests against recipient counts, this metric avoids population size bias 
and highlights borough-specific trends in law enforcement relative to assistance programs.


```python
# Step 1: Ensure 'Date' Column is in Monthly Format
# Convert the 'Date' column to a monthly period and back to a timestamp for consistent temporal grouping.
# This ensures all calculations are aligned to monthly intervals.
merged_df['Month'] = merged_df['Date'].dt.to_period('M').dt.to_timestamp()

# Step 2: Calculate Total Recipients
# Create a new column 'Total Recipients' as the sum of SNAP, Cash Assistance, and Medicaid recipients.
# This provides an aggregate view of individuals receiving any of these benefits in each borough and month.
merged_df['Total Recipients'] = (
    merged_df['Number of SNAP Benefit Recipients'] +
    merged_df['Number of Cash Assistance Recipients'] +
    merged_df['Medicaid Enrollees']
)

# Step 3: Remove Duplicate Rows
# If duplicate rows exist in the dataset, they could skew results during aggregation.
# Using `.drop_duplicates()` ensures each row is unique and avoids inflated counts or sums.
merged_df = merged_df.drop_duplicates()

# Step 4: Group by Borough and Month to Calculate Metrics
# Group the data by 'Borough' and 'Month', and calculate:
# - Total Recipients (sum): Aggregates the total number of recipients in each borough and month.
# - Total Arrests (count): Counts the number of rows (arrests) in each borough and month.
monthly_metrics = merged_df.groupby(['Borough', 'Month']).agg({
    'Total Recipients': 'sum',  # Sum all recipients across programs
    'Date': 'count'             # Count rows to calculate the number of arrests
}).rename(columns={'Date': 'Total Arrests'})  # Rename 'Date' column to 'Total Arrests'

# Step 5: Reset Index
# Reset the index to make 'Borough' and 'Month' accessible as columns for further analysis.
monthly_metrics = monthly_metrics.reset_index()

# Step 6: Calculate Arrests per Total Recipients
# Create a new column 'Arrests per Recipient' to measure the proportion of arrests to total recipients.
# This metric highlights the relationship between arrests and benefit recipients.
monthly_metrics['Arrests per Recipient'] = (
    monthly_metrics['Total Arrests'] / monthly_metrics['Total Recipients']
)

# Step 7: Display Metrics Without Rounding
# Print the resulting DataFrame to inspect the calculated metrics without applying rounding.
print("Monthly Borough-Specific Metrics (Unrounded):")
monthly_metrics
```

### D) Create Data Visualizations
1. Visualize borough-specific metrics using clear charts.
2. Include titles, axis labels, and legends for clarity.
### Interpretation of Visualizations
The line charts visualize the evolution of 'Arrests per Recipient' over time for each borough. The following
insights can be derived:
- Boroughs with consistently higher or lower ratios suggest systematic differences in law enforcement or
  demographic factors.
- Temporal spikes or drops in ratios may correlate with external events, policy changes, or other influences.


```python
import matplotlib.pyplot as plt

# Step 1: Ensure Metric Calculation
# Before plotting, confirm that the 'Arrests per Recipient' column exists in the `monthly_metrics` DataFrame.
# If not, calculate it by dividing total arrests by total recipients.
if 'Arrests per Recipient' not in monthly_metrics.columns:
    monthly_metrics['Arrests per Recipient'] = (
        monthly_metrics['Total Arrests'] / monthly_metrics['Total Recipients']
    )

# Step 2: Plot Arrests per Recipient for Each Borough
# Iterate through each unique borough in the dataset to create individual line charts.
for borough in monthly_metrics['Borough'].unique():
    # Filter data for the current borough to focus on borough-specific trends
    subset = monthly_metrics[monthly_metrics['Borough'] == borough]
    
    # Step 3: Create the Plot
    # Use a line chart to visualize 'Arrests per Recipient' over time (by month).
    plt.figure(figsize=(12, 6))  # Set the figure size for better visibility
    plt.plot(
        subset['Month'],  # X-axis: Month
        subset['Arrests per Recipient'],  # Y-axis: Arrests per Recipient
        label='Arrests per Recipient',  # Legend label
        marker='o',  # Add markers at data points for clarity
        color='blue'  # Set line color to blue
    )

    # Step 4: Configure the Chart
    # Add a title, axis labels, and a grid for better readability.
    plt.title(f"Arrests per Recipient Over Time in {borough}", fontsize=14)
    plt.xlabel("Month", fontsize=12)
    plt.ylabel("Arrests per Recipient", fontsize=12)
    plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
    plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add gridlines to the y-axis
    plt.tight_layout()  # Ensure layout is well-spaced

    # Step 5: Display the Chart
    # Render the plot for the current borough.
    plt.show()
```
